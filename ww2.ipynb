8. Write a program to accept the name and basic salary (datatype : float) of an employee. Calculate the dearness allowance as 40 % of basic and house rent allowance as 20 % of basic. Display the gross salary to the user. (Hint: gross salary = basic salary + dearness allowance + house rent allowance)

```python

employe_name=input("Enter the name of the employ")
print(" Name of the employ :",employe_name)
basic_salary=float(input("The basic salary of employ :"))
dearness_allowance=basic_salary*(40/100)
print("dearness_allowance is :",dearness_allowance)
house_rent_allowance=basic_salary*(20/100)
print("house_rent_allowance is :",house_rent_allowance)
gross_salary=house_rent_allowance+basic_salary+dearness_allowance
print("gross_salary is :",gross_salary)
```
